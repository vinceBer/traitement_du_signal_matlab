```python
import numpy as np
import scipy.io.wavfile as wav
import sounddevice as sd
#pip install sounddevice
import matplotlib.pyplot as plt
from scipy.signal import spectrogram

# Charger un fichier audio
Fs, signal = wav.read('Diner.wav')

# Afficher la fréquence d'échantillonnage
print('Fréquence d\'échantillonnage :', Fs, 'Hz')

# Durée du signal en secondes
duree = len(signal) / Fs
print('Durée du signal :', duree, 'secondes')
```

```python
def robotisation(signal, Fs) :
    signal_robotise = rob(signal, Fs)
    plot(signal_robotise, Fs)
    play(signal_robotise, Fs)
```

```python
def rob(y,Fs):
    Fc = 100  # par exemple
    # Modulation du signal avec une exponentielle complexe
    temps = np.arange(len(signal)) / Fs
    signal_modifie = signal[:, 0] * np.exp(-1j * 2 * np.pi * Fc * temps)

    # Extraire la partie réelle du signal modifié
    signal_robotise = np.real(signal_modifie)
    return signal_robotise
```

```python
def plot(signal, Fs): 
    # pour le spectrogramme
    window = 128  # Taille de la fenêtre pour le spectrogramme (réduite)
    overlap = 20  # Recouvrement entre les fenêtres
    nfft = 128  # Points de la FFT pour chaque segment
    
    N = len(signal)
    t = np.arange(N) / Fs
    f = np.arange(N) * Fs / N

    # Créer une fenêtre de Hanning 1-D de la taille spécifiée
    window = np.hanning(nfft)

    # Courbe de l'augmentation
    plt.figure(figsize=(8, 6))
    plt.subplot(4, 1, 1)
    plt.plot(t, signal)
    plt.xlabel('Temps (s)')
    plt.ylabel('Amplitude')
    plt.title('Évolution au cours du temps')

    # Tracer le spectre
    plt.subplot(4, 1, 2)
    Y = np.fft.fftshift(np.abs((1/N) * np.fft.fft(signal)))
    plt.plot(f - Fs / 2, Y)
    plt.xlabel('Fréquence (Hz)')
    plt.ylabel('Amplitude')
    plt.title('Spectre')

    plt.subplot(4, 1, 3)
    frequencies, times, Sxx = spectrogram(signal, fs=Fs, window=window, noverlap=overlap, nperseg=nfft, scaling='spectrum')
    plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx), shading='gouraud')
    plt.colorbar(label='dB')
    plt.ylabel('Fréquence (Hz)')
    plt.xlabel('Temps (s)')
    plt.title('Spectrogramme')
    plt.tight_layout()
    plt.show()


```

```python
def play(signal,Fs):
    # Jouer le signal modifié
    sd.play(signal_robotise, Fs)

    # Attendre jusqu'à ce que le signal ait fini de jouer
    sd.wait()
```

```python
# Utilisation de la fonction plot avec votre signal et votre fréquence d'échantillonnage
plot(signal_robotise, Fs)
```
